```python
#read in the file and make dictionaries of DBH at each census
in_file = open("/home/eeb177-student/Desktop/eeb-177/lab-work/exercise-7/cocoli.txt")
in_file.close = {}
```

```python
size_94 = {}
size_97 = {}
size_98 = {}
```

```python
#get rid of 
in_file.readline()

#
for line in in_file:
    items = line.split()
    
    #retrive relevant columns
    tree_id = items [0]
    s94 = float(items [4])
    s97 = float(items [5])
    s98 = float(items [6])
    
    #populate our dictionaries
    size_94[tree_id] = s94
    size_97[tree_id] = s97
    size_98[tree_id] = s98
    in_file.close()
    
```
